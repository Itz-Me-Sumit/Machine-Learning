### Data Modification

$$
h_w(x) = w_1 x_1 + w_2 x_2 + \dots + w_d x_d = x^T w
$$

### Definitions

- $x_i$: $i^{th}$ feature  
- $x$: Feature matrix of shape $(d, n)$  
- $w = [w_1, w_2, \dots, w_d]^T$: Weight vector  

---

The above model **always passes through the origin**.

To fix this, add an intercept term $w_0$ with a dummy feature $x_0 = 1$.

$$
h_w(x) = w_0 x_0 + w_1 x_1 + \dots + w_d x_d = x^T w
$$

---

- $x_0$ is the **dummy feature**
- Set $x_0 = 1$ for all examples

---

- $w$: shape $(d+1, 1)$  
- $x$: shape $(d+1, n)$  
- First row of $x$ → all 1s  

---

## Task

- Add the dummy feature to **Training_data** and **test_data**
- Use updated matrices for model training

**Note:** You may convert training data shape to $(d, n)$ if convenient.

```python
# For Training_data
train_ones_array = np.ones([1 , Training_data.shape[0]])
# print(train_ones_array, "\n \n" ,len(train_ones_array))
print("Ones Array Shape: ",train_ones_array.shape)

print("Training Data Previous Shape: ",Training_data.shape)
Training_data = Training_data.T
print("Training Data Shape After Transpose: ",Training_data.shape)
x_train = np.vstack([train_ones_array , Training_data])
print("Training Data Shape After Stack Ones_array above it: ",x_train.shape)

```

```python
# For Test_Data
test_ones_array = np.ones([1 , Test_data.shape[0]])
print("Ones Array Shape: " , test_ones_array.shape)

print("Shape of Test Data Before Transpose: ", Test_data.shape)
Test_data = Test_data.T
print("Shape of Test_Data after transpose: ",Test_data.shape)

x_test = np.vstack([test_ones_array , Test_data])
print("Shape Of test Data after adding Ones Row: ", x_test.shape)
```

### Question 4

If the solution of optimization problem is obtained by setting the first derivative of loss function (squared loss) to zero, find the value of $w_0 + w_1 + \dots + w_d$.

---

$$
\hat{y} = w_0 + w_1 x_1 + w_2 x_2 + \dots + w_d x_d
$$

### Yahan:

- $x_1, x_2, \dots, x_d$ → features (crime rate, rooms, etc.)
- $w_0, w_1, \dots, w_d$ → weights jo hume find karne hain
- $w_0$ → bias term (intercept)

---

$$
L(w) = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

$$
L(w) = (y - Xw)^T (y - Xw)
$$

---

$$
\frac{\partial L}{\partial w} = 0
$$

$$
-2 X^T (y - Xw) = 0
$$

$$
X^T y = X^T X w
$$

$$
w = (X^T X)^{-1} X^T y
$$

---

$$
w_0 + w_1 + \dots + w_d = \mathbf{1}^T w
$$

$$
= \mathbf{1}^T (X^T X)^{-1} X^T y
$$

```python
X = x_train.T
Y = labels
print(X.shape , Y.shape)

XtX = X.T @ X
XtY = X.T @ Y

W = np.linalg.inv(XtX) @ XtY
print(W.shape)

Sum = np.sum(W)
print(Sum)

```

- Question 5
    - Find the average of the predictions made by the above model.

```python
print(W.shape , x_train.shape)

prediction = W.T @ x_train
print(prediction.shape)
prediction_avg = np.mean(prediction)
print(prediction_avg)
```

### Question 6

Find the loss for the training data points using the above model.  
Consider the loss to be defined as:

$$
\sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
$$

Where $\hat{y}_i$ is the prediction for the $i^{th}$ data point.

```python
prediction = W.T @ x_train
error = labels - prediction
error_sq = error**2
MSE = np.mean(error_sq)
Rote_MSE = np.sqrt(MSE)
print(Rote_MSE)
```

### Question 8

Find the weights using gradient descent.  
Use a constant learning rate of $\eta = 10^{-10}$.

- Initialize the weight vector as a zero vector  
- Update the weights for 100 iterations  

Enter the **sum of all the weights**.

### Solution

### Gradient Descent Formula

$$
W \leftarrow W - 2\eta \left[(XX^T)W - XY\right]
$$

---


### Step by step solution:

**Step 1 — Initialize:**

$$
W =
\begin{bmatrix}
0 \\
0 \\
\vdots \\
0
\end{bmatrix}
\quad \text{shape: (14,)}
$$

---

**Step 2 — Gradient calculate karo:**

$$
\text{gradient} = (XX^T)W - XY
$$

- $(XX^T)W \rightarrow (14,14) \times (14,) = (14,)$  
- $XY \rightarrow (14,404) \times (404,) = (14,)$  

Final gradient shape: $(14,)$

---

**Step 3 — Update karo:**

$$
W \leftarrow W - 2\eta \cdot \text{gradient}
$$

---

**Step 4 — 100 baar repeat karo**

---

**Step 5 — Answer:**

$$
\text{ans} = \sum_{i=0}^{d} w_i = w_0 + w_1 + \dots + w_{13}
$$

```python
X = x_train
print(X.shape)
Y = labels
print(Y.shape)
eta = 1e-10
W = np.zeros(X.shape[0])
print(W.shape)

for i in range(100):
    gradient = ( X @ X.T ) @ W - X@Y
    W = W - 2*eta*(gradient)
Sum = np.sum(W)
print(Sum)
```
