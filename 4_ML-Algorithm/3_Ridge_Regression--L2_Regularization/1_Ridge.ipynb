# Ridge (L2 Regularization)

In Ridge regression, we modify the original Mean Squared Error (MSE)
by adding an L2 penalty term.

---

## Original Cost Function (MSE)

$$
\text{MSE}(w) = \frac{1}{n} \sum_{i=1}^{n} (y_i - x_i^T w)^2
$$

---

## New Cost Function (Ridge)

$$
J(w) = \text{MSE}(w) + \lambda \sum_{j=1}^{d} w_j^2
$$

---

## In Compact Vector Form

$$
J(w) = \frac{1}{n} \| y - Xw \|_2^2 + \lambda \| w \|_2^2
$$

where

$$
\| w \|_2^2 = \sum_{j=1}^{d} w_j^2
$$

---

## Meaning of Each Term

- First term → Data fitting (Mean Squared Error)
- Second term → L2 penalty (sum of squared weights)
- $\lambda \ge 0$ → regularization strength

If $\lambda$ increases:
- All coefficients shrink smoothly
- Rarely become exactly zero
- Model becomes more stable
