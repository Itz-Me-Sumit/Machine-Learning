# Lasso (L1 Regularization)

In Lasso regression, we modify the original Mean Squared Error (MSE) cost function
by adding an L1 penalty term.

---

## Original Cost Function (MSE)

$$
\text{MSE}(w) = \frac{1}{n} \sum_{i=1}^{n} (y_i - x_i^T w)^2
$$

---

## New Cost Function (Lasso)

$$
J(w) = \text{MSE}(w) + \lambda \sum_{j=1}^{d} |w_j|
$$

---

## In Compact Vector Form

$$
J(w) = \frac{1}{n} \| y - Xw \|_2^2 + \lambda \| w \|_1
$$

where

$$
\| w \|_1 = \sum_{j=1}^{d} |w_j|
$$

---

## Meaning of Each Term

- First term → Data fitting (Mean Squared Error)
- Second term → L1 penalty (sum of absolute values of weights)
- $\lambda \ge 0$ → regularization strength

If $\lambda$ increases:
- More coefficients shrink
- Some become exactly zero
- Model becomes sparse
