```python
"""
.plot() , .xlabel() & .ylabel() , 
"""


"""
plt.plot(x,y, color="color_name" , linestyle="line_style" , linewidth="line_width" , marker="marker_symbol" , label="label_name")
"""

import matplotlib.pyplot as plt
months = [1,2,3,4]
sales = [1000 , 1500 , 1200 , 1800]
plt.plot(months , sales , color='red' , linestyle='--' , linewidth=2 , marker = 'o' , label='2025 Sales Data')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

- legend()

```python
"""
plt.legend(loc = "upper left" OR "lower right")
plt.grid()
"""


import matplotlib.pyplot as plt
months = [1,2,3,4]
sales = [1000 , 1500 , 1200 , 1800]
plt.plot(months , sales , color='red' , linestyle='--' , linewidth=2 , marker = 'o' , label='2025 Sales Data')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(loc='upper left' , fontsize=12) # it'll Show label written inside there.
plt.grid(color='gray' , linestyle=':' , linewidth=1)
plt.show()
```

```python
"""
plt.xlim(start , end)
plt.ylim(start , end)
"""


import matplotlib.pyplot as plt
months = [1,2,3,4]
sales = [1000 , 1500 , 1200 , 1800]
plt.plot(months , sales , color='red' , linestyle='--' , linewidth=2 , marker = 'o' , label='2025 Sales Data')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(loc='upper left' , fontsize=12) # it'll Show label written inside there.
plt.grid(color='gray' , linestyle=':' , linewidth=1)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.show()
```

```python
"""
Axis Ticks :-
    - plt.xticks()
    - plt.ytikes()
"""

import matplotlib.pyplot as plt
months = [1,2,3,4]
sales = [1000 , 1500 , 1200 , 1800]
plt.plot(months , sales , color='red' , linestyle='--' , linewidth=2 , marker = 'o' , label='2025 Sales Data')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(loc='upper left' , fontsize=12) # it'll Show label written inside there.
plt.grid(color='gray' , linestyle=':' , linewidth=1)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4] , ['M1' , 'M2' , 'M3' , 'M4'])
plt.show()
```
