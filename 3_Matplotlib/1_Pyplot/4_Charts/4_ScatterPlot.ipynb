```python
"""
    plt.scatter(x,y , color='color_name' , marker='marker_style' , label='label_name')
"""
import matplotlib.pyplot as plt
hours_studied = [1,2,3,4,5,6,7,8]
exams_scores = [50,55,60,65,70,75,80,85]
#scatter_plot
plt.scatter(hours_studied , exams_scores , color='red' , marker='o' , label='performance Vs Hustle')
plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.title("Relationship between Hours Studied and Marks Gained")
plt.legend()
plt.grid(True)
plt.show()
```
